```python
import requests
import tabula
import pandas as pd
from xml.etree import ElementTree as ET

# https://estated.com/
# https://estated.com/developers/docs/v4/property/overview#request
# Download the PDF file
url = "https://ttc.lacounty.gov/wp-content/uploads/2023/03/2023A-auction-property-list.pdf"
response = requests.get(url)

# Write the PDF file to disk
with open("2023A-auction-property-list.pdf", "wb") as f:
    f.write(response.content)

# Extract tables from the PDF file
tables = tabula.read_pdf("2023A-auction-property-list.pdf", pages="all")

print('Type:', type(tables))
print('Table:', tables)

# Sort the tables based on a column
for table in tables:
    sorted_table = table.sort_values(by="MIN BID", ascending=True)
    print(sorted_table)
```
